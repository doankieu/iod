```python
import pandas as pd
import numpy as np
import seaborn as sns                       #visualisation
import matplotlib.pyplot as plt             #visualisation
%matplotlib inline     
sns.set(color_codes=True)
```

```python
df = pd.read_csv('data.csv')
```

```python
df = df.drop_duplicates()
```

```python
sns.boxplot(x=df['MSRP'])
```

```python
df.Make.value_counts().nlargest(40).plot(kind='bar', figsize=(10,5))
plt.title("Number of cars by make")
plt.ylabel('Number of cars')
plt.xlabel('Make');
```

```python
plt.figure(figsize=(10,5))
c= df.corr()
sns.heatmap(c,cmap="BrBG",annot=True)
c
```

```python
df_sns_chart = df.drop(['Year','highway MPG'],axis = 1)
```

```python
sns.set(rc={'figure.figsize':(16,10)})
#droping irrelevant feature to make pairplot less noisy. it is noticeable that I am dropping MPG features, as I have
#to double check these features relevancy to Electric cars
df_electric_sns = df_electric.drop(['Year','Engine Fuel Type','Engine Cylinders','highway MPG','city mpg'],axis =1)
sns.pairplot(df_electric_sns, hue= 'Make')
```

<div style="color:green"><h2>Tesla cars are a bit pricy. BMW & Ford rank high in Popularity. However, I am trying not reading to much into Popularity. Here is my reason:</h2><div>

```python
sns.set(rc={'figure.figsize':(15,10)})
sns.barplot(x = df['Popularity'], y = df['Make'])
```

```python
toyota_pop = df[df['Make']=='Toyota']['Popularity'].nunique()
BMW_pop = df[df['Make']=='BMW']['Popularity'].nunique()
Subaru_pop = df[df['Make']=='Subaru']['Popularity'].nunique()
print(toyota_pop,BMW_pop,Subaru_pop)
```

<div style="color:green"><h2> The popularity feature is biased toward brands. The feature disregards other features such as models, years. So I think we have to double check with the data collecter to see how it is measured<h2></div>
